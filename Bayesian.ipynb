```python
df_total = df_long.groupby(["Year_Index", "Age_Index", "Age_Bin", "Year", "Cohort_Index"]).Population.sum().reset_index()
```

```python
age_idx = df_total["Age_Index"].values
year_idx = df_total["Year_Index"].values
cohort_idx = df_total["Cohort_Index"].values
pop_obs = df_total["Population"].values

n_ages = df_total["Age_Index"].nunique()
n_cohorts = maxdf_total["Cohort_Index"].nunique()
```

```python
pop_scaled = pop_obs / 1000.0
log_offset = np.log(pop_scaled.mean()) 
with pm.Model() as apc_model:
    age_effect = pm.GaussianRandomWalk("age_effect", sigma =0.05, shape = n_ages)
    period_effect = pm.GaussianRandomWalk("period_effect", sigma = 0.05, shape = 3)
    cohort_effect = pm.GaussianRandomWalk("cohort_effect", sigma =0.05, shape = n_cohorts)
    intercept = pm.Normal("intercept", mu = log_offset, sigma = 0.5)
    log_mu = intercept + age_effect[age_idx] + period_effect[year_idx] + cohort_effect[cohort_idx]
    y = pm.Poisson("y", mu=pm.math.exp(log_mu), observed=pop_scaled)
    trace = pm.sample(1000, tune = 1000, target_accept = 0.95, init = "adapt_diag")
```

```python
geos = df_long["geography"].unique()
ages = range(n_ages)

forecast_data = []
for g in geos:
    for a in ages:
        forecast_data.append({"geography":g, "Age_Index":a, "Year_Index":2, "Cohort_Index":(2*2)- a +25 })
df_forecast_input = pd.DataFrame(forecast_data)
```

```python
with apc_model:
    pm.set_data({"age_idx": df_forecast_input["Age_Index"].values, 
                 "year_idx":df_forecast_input["Year_Index"].values, 
                 "cohort_idx": df_forecast_input["Cohort_Index"].values})
    
    post_pred = pm.sample_posterior_predictive(trace)
    
prediction = post_pred.posterior_predictive["y"].mean(axis=(0,1))
df_forecast_input["Population_2031"] = prediction * 1000

    
```

```python
import pymc as pm
import numpy as np
import pandas as pd

# 1. Aggregate counties to get a National Total
df_total = df_long.groupby(['Year_Index', 'Age_Index', 'Age_Bin', 'Year', 'Cohort_Index']).Population.sum().reset_index()

# 2. Setup Data for Model (Scaling by 1000 for stability)
pop_obs = df_total['Population'].values / 1000.0
age_idx = df_total['Age_Index'].values
year_idx = df_total['Year_Index'].values
cohort_idx = df_total['Cohort_Index'].values

n_ages = df_total['Age_Index'].nunique()
# We calculate n_cohorts based on the max index used in the whole set
n_cohorts = max(df_total['Cohort_Index'].max(), (2*2 - 0 + 25)) + 1 

with pm.Model() as fast_model:
    # Data containers (so we can swap them for forecasting)
    a_in = pm.Data("a_in", age_idx)
    y_in = pm.Data("y_in", year_idx)
    c_in = pm.Data("c_in", cohort_idx)

    # Priors
    alpha = pm.GaussianRandomWalk("age_effect", sigma=0.05, shape=n_ages)
    beta = pm.GaussianRandomWalk("period_effect", sigma=0.05, shape=3) # 2011, 2021, 2031
    gamma = pm.GaussianRandomWalk("cohort_effect", sigma=0.05, shape=n_cohorts)
    intercept = pm.Normal("intercept", mu=np.log(pop_obs.mean()), sigma=0.5)

    # The Formula
    log_mu = intercept + alpha[a_in] + beta[y_in] + gamma[c_in]
    
    # Likelihood
    y_obs = pm.Poisson("y_obs", mu=pm.math.exp(log_mu), observed=pop_obs)

    # Sampling (This will be very fast now!)
    trace = pm.sample(1000, tune=1000, target_accept=0.9)
```
